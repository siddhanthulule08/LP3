```python
import numpy as np
```

```python
def f(x):
    return (x + 3)**2

def f_derivative(x):
    return 2 * (x + 3)
```

```python
x_current = 2  # Starting point
learning_rate = 0.1  # Learning rate
n_iterations = 100
```

```python
for i in range(n_iterations):
    gradient = f_derivative(x_current)  # Compute the gradient
    x_current = x_current - learning_rate * gradient  # Update x based on gradient

    # Optional: Print progress
    # print(f"Iteration {i+1}: x = {x_current:.4f}, f(x) = {f(x_current):.4f}")
```

```python
print("\nLocal minimum occurs at x =", x_current)
print("Minimum value of function:", f(x_current))
```
